# **3 - Data Cleaning**

## Importing Libraries

```python
import pandas as pd
import pickle
```

## Importing Pickles

```python
with open('fighters.pkl', 'rb') as f:
    fighters = pickle.load(f)
with open('fights.pkl', 'rb') as f:
    fights = pickle.load(f)
```

## Dropping Columns

```python
# Dropping Post-Fight Columns
fights = fights[['R_fighter', 'B_fighter', 'Format', 'Referee', 'date', 'location', 'Fight_type', 'Winner']]
```

## Renaming Columns

```python
fights.columns = ['R', 'B', 'Format', 'Referee', 'Date', 'Location', 'Division', 'Winner']
fighters.columns = ['Name', 'Height', 'Weight', 'Reach', 'Stance', 'Birthday', 'SSLPM', 'SAC', 'SSAPM', 'SD', 'TAV', 'TAC', 'TD', 'SAV']

# SSLPM = Significant Strikes Landed Per Minute
# SAC = Striking Accuracy
# SSAPM = Significant Strikes Absorbed Per Minute
# SD = Striking Defense
# TAV = Takedowns Average
# TAC = Takedown Accuracy
# TD = Takedown Defense
# SAV = Submission Average

print('Fighters')
print('************************************')
display(fighters.head())
print('Fights')
print('************************************')
display(fights.head())
```

## Merging DataFrames

```python
df = pd.merge(fights, fighters, left_on = 'R', right_on = 'Name', how = 'left')
df = pd.merge(df, fighters, left_on = 'B', right_on = 'Name', how = 'left')
# Renaming Columns
df.columns = ['R', 'B', 'Format', 'Referee', 'Date', 'Location', 'Division', 'Winner', 'R_Name', 'R_Height', 'R_Weight', 'R_Reach', 'R_Stance', 'R_Birthday', 'R_SSLPM', 'R_SAC', 'R_SSAPM', 'R_SD', 'R_TAV', 'R_TAC', 'R_TD', 'R_SAV', 'B_Name', 'B_Height', 'B_Weight', 'B_Reach', 'B_Stance', 'B_Birthday', 'B_SSLPM', 'B_SAC', 'B_SSAPM', 'B_SD', 'B_TAV', 'B_TAC', 'B_TD', 'B_SAV']
print('Fighters and Fights')
display(df.info())
print('------------------------------------')
print('Head')
display(df.head())
print('------------------------------------')
print('Tail')
display(df.tail())
```

## Modifying Columns

### Converting Height, Weight, and Reach

```python
# Convert from FT IN to CM
def convert_height(height):
    if pd.isna(height):
        return np.nan
    feet, inches = height.split(" ")
    feet = float(feet[0])
    inches = float(inches[0])
    return (feet * 30.48) + (inches * 2.54)

# Convert from LB to KG
def convert_weight(weight):
    if pd.isna(weight):
        return np.nan
    weight = weight.split(" ")
    weight = float(weight[0])
    return weight * 0.4536

# Convert from IN to CM
def convert_reach(reach):
    if pd.isna(reach):
        return np.nan
    reach = float(reach[0:2])
    return reach * 2.54

df['R_Height'] = df['R_Height'].apply(convert_height)
df['R_Weight'] = df['R_Weight'].apply(convert_weight)
df['R_Reach'] = df['R_Reach'].apply(convert_reach)
df['B_Height'] = df['B_Height'].apply(convert_height)
df['B_Weight'] = df['B_Weight'].apply(convert_weight)
df['B_Reach'] = df['B_Reach'].apply(convert_reach)

display(df[['R_Height', 'R_Weight', 'R_Reach', 'B_Height', 'B_Weight', 'B_Reach']].head())
```

### Calculating Age

```python
# Convert to DateTime
df['Date'] = pd.to_datetime(df['Date'])
df['R_Birthday'] = pd.to_datetime(df['R_Birthday'])
df['B_Birthday'] = pd.to_datetime(df['B_Birthday'])
# Age in Days
df['R_Age'] = (df['Date'] - df['R_Birthday']).dt.days
df['B_Age'] = (df['Date'] - df['B_Birthday']).dt.days

display(df[['R_Age', 'B_Age']].head())
```

### Binarize Winner

```python
# Red (R) Wins = 1, Blue (B) Wins = 0
def binarize_winner(row):
    if row['Winner'] == row['R']:
        return 1
    else:
        return 0
df['Winner'] = df.apply(binarize_winner, axis = 1)
display(df['Winner'].head())
```

## Handle Missing Values

```python
# Imputing Reach

print('Before')
print('------------------------------------')
for column in df.columns:
    print(column, 'Missing Values:', df[column].isnull().sum())

df['R_Reach'] = df['R_Reach'].fillna(df['R_Height'])
df['B_Reach'] = df['B_Reach'].fillna(df['B_Height'])

print('\n')
print('After')
print('------------------------------------')
for column in df.columns:
    print(column, 'Missing Values:', df[column].isnull().sum())
```

## Adding Columns

```python
df['Fight_Name'] = df['R_Name'] + ' vs. ' + df['B_Name']
```

## Changing Data Types

```python
# Before
print(df.info())

# Converting Percentages to Floats
cols = ['R_SAC', 'R_SD', 'R_TAC', 'R_TD', 'R_SAC', 'B_SAC', 'B_SD', 'B_TAC', 'B_TD']
for col in cols:
    df[col] = df[col].astype(str).str.rstrip('%').astype('float') / 100.0

# Converting into Floats, Objects (Strings), Datetime, and Categories
dtype_dict = {
    'R': 'object', 'B': 'object', 'Format': 'category', 'Referee': 'object', 'Date': 'datetime64', 'Location': 'object', 'Division': 'category', 'Winner': 'int64',
    'R_Name': 'object', 'R_Height': 'float64', 'R_Weight': 'float64', 'R_Reach': 'float64', 'R_Stance': 'category',  'R_SSLPM': 'float64', 'R_SAC': 'float64', 'R_SSAPM': 'float64', 'R_SD': 'float64', 'R_TAV': 'float64', 'R_TAC': 'float64', 'R_TD': 'float64', 'R_SAV': 'float64',
    'B_Name': 'object', 'B_Height': 'float64', 'B_Weight': 'float64', 'B_Reach': 'float64', 'B_Stance': 'category', 'B_SSLPM': 'float64', 'B_SAC': 'float64', 'B_SSAPM': 'float64', 'B_SD': 'float64', 'B_TAV': 'float64', 'B_TAC': 'float64', 'B_TD': 'float64', 'B_SAV': 'float64',
    'Fight_Name': 'object'
}

df = df.astype(dtype_dict)
# After
print(df.info())
```

## Re-Ordering Columns


```python
order = ['Fight_Name', 'Date', 'Location', 'R', 'B', 'Winner', 'Division', 'Format', 'Referee',  
        'R_Name', 'R_Age', 'R_Height', 'R_Weight', 'R_Reach', 'R_Stance', 'R_Birthday', 'R_SSLPM', 'R_SAC', 'R_SSAPM', 'R_SD', 'R_TAV', 'R_TAC', 'R_TD','R_SAV',
        'B_Name', 'B_Age', 'B_Height', 'B_Weight', 'B_Reach', 'B_Stance', 'B_Birthday', 'B_SSLPM', 'B_SAC', 'B_SSAPM', 'B_SD', 'B_TAV', 'B_TAC', 'B_TD', 'B_SAV']

df = df.reindex(columns = order)

df.head()
```
